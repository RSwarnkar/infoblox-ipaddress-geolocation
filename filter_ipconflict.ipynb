```python
import numpy as np
import pandas as pd
import radix as rx
```

```python
df = pd.read_csv("output09-gnoc_networks-final_with_ipconflict-02.csv", encoding='utf-8')
```

```python
df.head()
```

```python
df.columns
```

```python
root_node = rx.Radix()
```

```python
for i in range(0, len(df)):
    
    SerialNumber      = str(df.iloc[i]['SerialNumber']).strip('\n')
    NetworkIPv4_CIDR  = str(df.iloc[i]['NetworkIPv4_CIDR']).strip('\n')
    CountryName       = str(df.iloc[i]['CountryName']).strip('\n')
    CountryCode       = str(df.iloc[i]['CountryCode']).strip('\n')
    RegionName        = str(df.iloc[i]['RegionName']).strip('\n')
    RegionCode        = str(df.iloc[i]['RegionCode']).strip('\n')
    CityName          = str(df.iloc[i]['CityName']).strip('\n')
    Latitude          = str(df.iloc[i]['Latitude']).strip('\n')
    Longitude         = str(df.iloc[i]['Longitude']).strip('\n')
    
    rn = root_node.add(NetworkIPv4_CIDR)
    rn.data["SerialNumber"] = SerialNumber
    rn.data["CountryName"] = CountryName
    rn.data["CountryCode"] = CountryCode
    rn.data["RegionName"] = RegionName
    rn.data["RegionCode"] = RegionCode
    rn.data["CityName"] = CityName
    rn.data["Latitude"] = Latitude
    rn.data["Longitude"] = Longitude
```

```python
output_file = open("output2.csv","w")
```

```python
nodes = root_node.nodes()

line = "NetworkIPv4_CIDR;SerialNumber;CountryName;CountryCode;RegionName;RegionCode;CityName;Latitude;Longitude\n"
output_file.write(line)


for rnode in nodes:
    line = str(rnode.prefix)+";"+ \
    rnode.data["SerialNumber"]+";"+ \
    rnode.data["CountryName"]+";"+ \
    rnode.data["CountryCode"]+";"+ \
    rnode.data["RegionName"]+";"+ \
    rnode.data["RegionCode"]+";"+ \
    rnode.data["CityName"]+";"+ \
    rnode.data["Latitude"]+";"+ \
    rnode.data["Longitude"]+"\n"
    
    output_file.write(line)
        
output_file.close()
```
